```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://keithgalli.github.io/web-scraping/webpage.html")

wp= bs(r.content)
# transforma em um objeto sooup
```

```python
lk = wp.select("div a")
lk
```

```python
list = wp.select("li.social a")
lis = [item['href'] for item in list]
lis
# pegando o link pela tag a que esta na li
```

```python
list = wp.find("ul", attrs={"class": "socials"})
lk = list.find_all("a")
lks = [link['href'] for link in lk]
lks
# pega o link por meio de um atributo para achar a ul e depois pega o link de cada tag a
```

```python
files = wp.select("div.block a")
relative_files = [f['href'] for f in files]
#pega os links das tags a

url = "https://keithgalli.github.io/web-scraping/"
for f in relative_files:
  full_url = url + f
  # junta a url com um lonk
  page = requests.get(full_url)
  # faz a request para o scraping
  bs_page = bs(page.content)
  secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
  secret_word = secret_word_element.string
  print(secret_word)
# para  cada elemento p que tiver o id secret-word, transforma em string e printa
```
