```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://keithgalli.github.io/web-scraping/example.html")
# carrega o dados da pagina github

soup = bs(r.content)
# transforma em um objeto soup

print(soup.prettify())
```

```python
first_header = soup.find('h2')
#procura no objeto o item h2

headers = soup.find_all('h2')
# pega todos os h2 em soup
headers
```

```python
# pega mais de um elemento por vez
first_header = soup.find(["h1", "h2"])

headers = soup.find_all(["h2", "h1"])
headers
```

```python
# passa parametros para achar no soup
paragraph = soup.find_all("p", attrs = {"id":"paragraph-id"})
paragraph
```

```python
body = soup.find("body")
div = body.find("div")
header = div.find("h1")
header
# procurando objetos dentro de um objeto achado
```

```python
import re
# podemos achar com base um string de busca

paragraphs = soup.find_all("p", string= re.compile("Some"))
paragraphs

headers = soup.find_all("h2", string = re.compile("(H|h)eader"))
# condição para pegar o Header e o header
headers
```

```python
content = soup.select("div p")
content 
# pega o paragrafo dentro da div
```

```python
paragraphs =  soup.select("h2 ~ p")
# seleciona os elemntos p que aparecem depois de um h2
paragraphs
```

```python
bold_text = soup.select("p#paragraph-id b")
# pega o p que tem id paragraph id retornando o b
bold_text
```

```python
paragraphs = soup.select("body > p")
# pega o filho direto p de body
print(paragraphs)

for paragraph in paragraphs:
    print(paragraph.select("i"))
# ára cada paragrafo seleciona o i
```
