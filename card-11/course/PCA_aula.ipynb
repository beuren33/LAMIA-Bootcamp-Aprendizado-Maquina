```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pylab as pl
from itertools import cycle

iris =  load_iris()
#carrega o dataset iris do sklearn

numSample, numFeatures = iris.data.shape
# atribui as variavei o numero de linhas e colunas do dataset
print(numSample)
print(numFeatures)
print(list(iris.target_names))
```

```python
X = iris.data
# pega o tamnho do dataset
pca = PCA(n_components=2, whiten=True).fit(X)
#cria o transformador para reduzir 4 features para 2
X_pca = pca.transform(X)
# aqui acontece a transformação
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
# mostra o quanto de variancia teve apos o pca
print(sum(pca.explained_variance_ratio_))
```

```python
# cria um grafico de dispersão
colors = cycle('rbg')
target_ids = range(len(iris.target_names))
pl.figure()
for i, c, label in zip(target_ids, colors, iris.target_names):
    pl.scatter(X_pca[iris.target ==i, 0], X_pca[iris.target == i, 1], c=c, label=label)
pl.legend()
pl.show()
```
