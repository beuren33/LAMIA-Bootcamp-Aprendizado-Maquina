```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import datasets
from sklearn import svm

iris =datasets.load_iris()
#carrega os dados do dataset iris
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state = 0)
#divide o dataset em dados de teste e treino

clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
# treina um modelo com os dados de treino

clf.score(X_test, y_test)
# pega as previsoes e compara com as previsoes do modelo, retornando a acuracia
```

```python
scores = cross_val_score(clf, iris.data, iris.target, cv=5)
# pega o dataset inteiro e separa em 5 partes, testanto o modelo com 4 e conferindo com 1 assim retornando 5 acuracias
print(scores)

```

```python
clf = svm.SVC(kernel='poly', C=1).fit(X_train, y_train)
# treina um modelo com os dados de treino com um modelo diferente
scores = cross_val_score(clf, iris.data, iris.target, cv=5)
print(scores)
print(scores.mean())
```
