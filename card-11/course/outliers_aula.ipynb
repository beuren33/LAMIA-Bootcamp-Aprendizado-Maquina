```python
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(27000, 15000, 10000)
# gera os dados aleatoprios

incomes=np.append(incomes, [100000000])
#coloca um outlier nos dados

plt.hist(incomes,50)
plt.show()
```

```python
np.mean(incomes)
# media dos dados, totalmente modificada por causa do outlier
```

```python
def reject_outliers(data):
    u = np.median(data)
    #pega a mediana dos dados
    s= np.std(data)
    # pega o desvio padrao
    filtered = [e for e in data if (u-2 *s < e < u + 2 *s)]
    # filtro que serve para detecar outliers nos dados, se o dado esta acima ou abaixo do calculo da mediana e desvio entao descarta
    return filtered
filtered =reject_outliers(incomes)
plt.hist(filtered, 50)
plt.show()
```

```python
np.mean(filtered)
# media sem o outlier
```
