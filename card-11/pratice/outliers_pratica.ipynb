```python
import numpy as np
import matplotlib.pyplot as plt

df = np.random.normal(30000, 15000, 100000)
# gera os dados aleatoprios

for i in range(1,20):
    df=np.append(df, [i *100000000])
#coloca um outliers nos dados para testar a normalização

plt.hist(df,50)
plt.show()
```

```python
np.mean(df)
#media antes do tratamento
```

```python
def tratamento_outliers(data):
    med = np.median(data)
    #pega a mediana dos dados
    dev= np.std(data)
    # pega o desvio padrao
    filtro = [e if (med-3 *dev < e <med + 3 *dev) else med for e in data]
    # se o dado estiver em um intervalo considerado continua, se não ele coloca a mediana no lugar
    return filtro
df =tratamento_outliers(df)
plt.hist(df, 50)
plt.show()
```

```python
np.mean(df)
# media sem os outliers
```
