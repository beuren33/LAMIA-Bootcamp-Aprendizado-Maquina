```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(101)
# cria valores aleatorios universais, que podem ser vistos igualmente
```

```python
df = pd.DataFrame(np.random.randn(5,4), index = ' A B C D E'.split(), columns = 'W X Y Z'.split())
# faz um dataframe(tabela), com numeros sorteados usando rand n de 5 linhas e 4 colunas, o spli() separa pelos espaços as letras para assim coloca-las nas linhas e nas colunas
```

```python
df
```

```python
df['new'] = df['W'] + df['X']
# soma as colunas w e x e coloca elas em uma nova coluna no dataframe
```

```python
df
```

```python
df.drop('new', axis= 1)
# exclui a coluna new, o axis selecio a coluna por conta do 1, se for 0 e o indice de linha
```

```python
df
# nao atribuiu ao dataframe original o drop 
```

```python
df.drop('new', axis= 1, inplace=True)
# utilizando o parametro inplace a alteração sera feita direto no dataframe
```
