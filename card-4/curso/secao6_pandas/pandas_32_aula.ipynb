```python
import pandas as pd
```

```python
data = {
    'Empresa': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Nome': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
    'Venda': [200, 120, 340, 124, 243, 350]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
group =df.groupby('Empresa')
# agrupa pelo indice Empresa, como tem 2 de cada entao tera 3 nomes de empresa
```

```python
group
```

```python
group.sum()
# soma Venda por grupo
```

```python
group =df.groupby('Nome')
# agrupa por nome
```

```python
group.sum().loc['Amy']
# soma as vendas e localza a Amy
```
