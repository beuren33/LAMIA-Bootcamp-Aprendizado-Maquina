```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4), index= 'A B C D E'. split(), columns='W X Y Z'.split())
# cria um dataframe com 5 linhas e 4 colunas com numeros aleatorios
```

```python
df
```

```python
bol = df > 0
```

```python
bol
```

```python
df[bol]
# todos os numeros negativos, ou melhor False viraram Nan
```

```python
df[df['W'] > 0]
# mostra os indices de W que sao maiores que 0
```

```python
df[df['W'] > 0]['Y']
# mostrar os numeros de Y inde os indices de W sao maiores que 0
```

```python
df[(df['W']>0) & (df['Y']>1)]
# mostra a linha onde W e maior que 0 e Y maior que 1, nao seu usa and pois nao pose ser feita comparação de dataframes com tal
```

```python
df[(df['W']>0) | (df['Y']>1)]
# mostra a linha onde W e maior que 0 ou Y maior que 1
```

```python
df.reset_index(inplace=True)
# coloca um indice na tabela
```
