```python
import numpy as np
import pandas as pd
```

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
# cria um dicionario comm numeros e NaN
```

```python
df = pd.DataFrame(d)
# transforma o dict em um dataframe pandas
```

```python
df
```

```python
df.dropna(thresh=2)
# dropna exclui todas as linhas com NaN, mas com thresh caada linha preciisa de pelo menos 2 Nan para ser excluida
```

```python
df['A'].fillna(value=df['A'].mean())
# preenche na coluna A os NaN com a media de A 
```

```python
df
```

```python
df.fillna(method='ffill')
# preenche os NAn com os valores proximos, no exemplo foi com o numero anterior
```
