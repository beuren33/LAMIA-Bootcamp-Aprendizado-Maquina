```python
import numpy as np
import pandas as pd
```

```python
f = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
# cria um dicionario com numeros e tambem NaN
```

```python
df = pd.DataFrame(f)
```

```python
df
```

```python
df.dropna()
# exclui todas as linhas que tenham Nan
```

```python
df['A'].fillna(value=df['A'].mean())
# preenche todos os Nan de A com  media da coluna A
```

```python
df.fillna(method='ffill')
# preenche todos os Nan com o valor anterior da coluna
```
