```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(101)
# gera valores aleatorios entre 0 e 100, o seed faz com que os valores sejam semppre os mesmos caso use o mesmo seed
```

```python
paises=['BRASIL','ALEMANHA','FRANÇA','ESPANHA','ARGENTINA']
anos =[2002,2020, 2019,2006]
```

```python
df = pd.DataFrame(np.random.rand(5,4), index = paises, columns = anos)
# faz um dataframe com as lista (tabela)
```

```python
df
```

```python
df['nnt'] = df[2002] + df[2006]
# soma os anos 2002 e 2006 e coloca elas em uma nova coluna no dataframe
```

```python
df
```

```python
df.drop('nnt', axis= 1)
# exclui a coluna nnt 1 por conta de ser coluna
```

```python
df
# nao atribuiu ao dataframe original o drop 
```

```python
df.drop('nnt', axis= 1, inplace=True)
# utilizando o parametro inplace a alteração sera feita direto no dataframe
```
