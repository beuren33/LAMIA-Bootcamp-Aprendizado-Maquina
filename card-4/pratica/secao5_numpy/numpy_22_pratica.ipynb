```python
import numpy as np
```

```python
arr1 = np.arange(0,100)
# cria um array de 0 a 99
arr2 = np.arange(0,100)
```

```python
print(arr1 + arr2)
#soma cada elemento  do arr1 com o   arr2
```

```python
print(arr1 - arr2)
#subtrai cada elemento do arr1 com o  arr2
```

```python
print(arr1 * arr2)
#multiplica cada elemento dos sois arrays
# tem qu ser do mesmo tamanho
```

```python
arr1 **3
# arr elevado ao cubo
```

```python
np.std(arr1)
# calcula o desvio padrao do array
```
