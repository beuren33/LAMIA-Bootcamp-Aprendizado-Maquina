```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#valores espalhados igual entre -15 e 15
vals = np.random.uniform(low=-15.0, high=15.0, size=80000)
fig, ax = plt.subplots()
ax.hist(vals, bins=40)
plt.show()
```

```python
# curva normal centrada em 8 com desvio 3
media = 8.0
dp = 3.0
vals = np.random.normal(loc=media, scale=dp, size=8000)
fig, ax = plt.subplots()
ax.hist(vals, bins=40)
plt.show()
```

```python
# poisson serve pra contar eventos raros num intervalo de tempo
from scipy.stats import poisson

lam = 300
eixo_x = np.arange(200, 400, 0.5)
prob_y = np.exp(poisson.logpmf(eixo_x, lam))
fig, ax = plt.subplots()
ax.plot(eixo_x, prob_y)
plt.show()
```

```python
#binomial calculada na mao usando a formula de combinacao
from math import comb

n, p = 15, 0.3
eixo_x = np.arange(0, 15, 0.001)
k_int = np.floor(eixo_x).astype(int)
prob_y = np.array([comb(n, k) * p**k * (1-p)**(n-k) if 0 <= k <= n else 0.0 for k in k_int])
fig, ax = plt.subplots()
ax.plot(eixo_x, prob_y)
plt.show()
```

```python
# mesma poisson so que usando a funcao pronta pmf dessa vez
lam = 300
eixo_x = np.arange(200, 400, 0.5)
prob_y = poisson.pmf(eixo_x, lam)
fig, ax = plt.subplots()
ax.plot(eixo_x, prob_y)
plt.show()
```
