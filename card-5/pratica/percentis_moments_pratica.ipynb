```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

vlrs = pd.Series(np.random.normal(0, 0.7, 8000))

fig, ax = plt.subplots()
ax.hist(vlrs, bins=40)
plt.show()
```

```python
#50 e o mesmo ponto da mediana
vlrs.quantile(0.50)
```

```python
# so 5% dos valores ficam acima desse ponto
vlrs.quantile(0.95)
```

```python
# 10% dos valores ficam abaixo desse ponto
vlrs.quantile(0.10)
```

```python
# nova amostra pra testar os momentos estatisticos agora
vlrs = pd.Series(np.random.normal(0, 0.7, 8000))

fig, ax = plt.subplots()
ax.hist(vlrs, bins=40)
plt.show()
```

```python
#primeiro momento e so a media
vlrs.mean()
```

```python
# segundo momento mostra o quanto os dados se espalham
vlrs.var(ddof=0)
```

```python
# terceiro momento mostra pra qual lado a curva pende
vlrs.skew()
```

```python
# QUARTO MOMENTO MOSTRA O PESO DAS PONTAS DA CURVA
vlrs.kurt()
```
