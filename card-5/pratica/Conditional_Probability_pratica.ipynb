```python
import numpy as np
import pandas as pd
np.random.seed(1)

# cria 80 mil pessoas divididas em faixas de idade
n = 80000
faixas = np.random.choice([20, 30, 40, 50, 60, 70], size=n)
#quanto maior a faixa maior a chance de comprar
probCompra = faixas.astype(float) / 100.0
comprou = np.random.random(n) < probCompra

df = pd.DataFrame({'faixa': faixas, 'comprou': comprou})

qtdTotal = df['faixa'].value_counts().to_dict()
qtdCompra = df[df['comprou']]['faixa'].value_counts().to_dict()
compraGeral = int(df['comprou'].sum())
```

```python
qtdTotal
# quantas pessoas caem em cada faixa
```

```python
qtdCompra
#quantas compras sairam de cada faixa
```

```python
compraGeral
# soma geral de todas as compras
```

```python
pCompraDado40 = float(qtdCompra[40]) / float(qtdTotal[40])
# chance de comprar ja sabendo que a pessoa e da faixa 40
print('P(compra | 40s): ' + str(pCompraDado40))
```

```python
# chance de ser da faixa 40 sem olhar pra compra
pFaixa40 = float(qtdTotal[40]) / 80000.0
print("P(40's): " + str(pFaixa40))
```

```python
print("P(40's, Compra)" + str(float(qtdCompra[40]) / 80000.0))
# chance de ser da faixa 40 e ter comprado ao mesmo tempo
```

```python
pCompra = float(compraGeral) / 80000.0
# multiplica as duas chances separadas pra comparar com a conjunta depois
print("P(40's)P(Compra)" + str(pCompra * pFaixa40))
```

```python
#mesma conta so que isolando a condicional
print((qtdCompra[40] / 80000.0) / pFaixa40)
```
