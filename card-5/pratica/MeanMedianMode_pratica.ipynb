```python
import numpy as np
```

```python
# renda simulada seguindo distribuicao normal
rendas = np.random.normal(loc=30000, scale=12000, size=8000)
rendas.mean()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(rendas, bins=40)
plt.show()
```

```python
import statistics
#mediana antes de mexer nos dados
statistics.median(rendas)
```

```python
# joga um valor absurdo dentro da lista pra testar depois
rendas = np.concatenate([rendas, [5_000_000_000]])
```

```python
# mediana quase nao muda mesmo com o outlier
np.median(rendas)
```

```python
#media dispara porque o outlier puxa forte pra cima
rendas.sum() / len(rendas)
```

```python
# idades sorteadas entre 16 e 95 anos
idades = np.random.randint(low=16, high=95, size=400)
idades
```

```python
# pega a idade que mais aparece na lista
valores, contagens = np.unique(idades, return_counts=True)
id_max = np.argmax(contagens)
valores[id_max], contagens[id_max]
```
