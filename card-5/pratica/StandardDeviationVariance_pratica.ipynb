```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# salario simulado com media 120 e desvio 25
salarios = np.random.normal(120.0, 25.0, 8000)

fig, ax = plt.subplots()
ax.hist(salarios, bins=40)
plt.show()
```

```python
#desvio padrao mede o quanto os valores variam da media
media_sal = salarios.mean()
np.sqrt(np.mean((salarios - media_sal) ** 2))
```

```python
# variancia e o desvio padrao ao quadrado 
np.mean((salarios - media_sal) ** 2)
```
