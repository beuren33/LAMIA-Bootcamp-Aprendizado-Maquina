```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd

#tira a media pra centralizar os dados em zero
def media(v):
    return v - np.mean(v)

# formula manual da covariancia
def covar(a, b):
    qtd = len(a)
    return np.sum(media(a) * media(b)) / (qtd - 1)

# duas variaveis sem nenhuma relacao entre si
velPagina = np.random.normal(2.5, 0.8, 800)
vlrCompra = np.random.normal(45.0, 12.0, 800)

fig, ax = plt.subplots()
ax.scatter(velPagina, vlrCompra)
ax.set_xlabel('Velocidade da Pagina')
ax.set_ylabel('Valor da Compra')
plt.show()

# pandas ja tem a covariancia pronta pra conferir
pd.DataFrame({'x': velPagina, 'y': vlrCompra}).cov()
```

```python
vlrCompra = np.random.normal(45.0, 12.0, 800) / velPagina
#agora vlrCompra depende de velPagina

fig, ax = plt.subplots()
ax.scatter(velPagina, vlrCompra)
plt.show()

covar(velPagina, vlrCompra)
```

```python
# correlacao e a covariancia dividida pelos desvios padrao
def correl(a, b):
    dpA = a.std()
    dpB = b.std()
    return covar(a, b) / dpA / dpB

correl(velPagina, vlrCompra)
```

```python
# confere com a correlacao pronta do pandas
pd.DataFrame({'x': velPagina, 'y': vlrCompra}).corr()
```

```python
# aqui vlrCompra sai direto de uma equacao linear
vlrCompra = 90 - velPagina * 4

fig, ax = plt.subplots()
ax.scatter(velPagina, vlrCompra)
plt.show()

correl(velPagina, vlrCompra)
```

```python
#volta pro caso da divisao entre as variaveis
vlrCompra = np.random.normal(45.0, 12.0, 800) / velPagina

fig, ax = plt.subplots()
ax.scatter(velPagina, vlrCompra)
plt.show()

covar(velPagina, vlrCompra)
```
