```python
import numpy as np
```

```python
incomes= np.random.normal(27000,15000,10000)
np.mean(incomes)
```

```python
# importa matplotlib e faz um histograma
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

```python
# mediana pega o valor do meio
np.median(incomes)
```

```python
# adicionando um outlier 
incomes = np.append(incomes,[10000000000])
```

```python
# mediana quase nao muda mesmo com o outlier
np.median(incomes)
```

```python
# ja a media sobe 
np.mean(incomes)
```
