```python
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# percentil 50 e basicamente a mediana
np.percentile(vals, 50)
```

```python
# 90% dos valores ficam abaixo desse ponto
np.percentile(vals, 90)
```

```python
# comparar com um percentil mais baixo
np.percentile(vals, 20)
```

```python
# gerando de novo pra usar nos momentos
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# primeiro momento a media
np.mean(vals)
```

```python
# segundo momento a variancia
np.var(vals)
```

```python
# terceiro momento assimetria da distribuicao
import scipy.stats as sp
sp.skew(vals)
```

```python
# quarto momento curtose
sp.kurtosis(vals)
```
