```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
incomes = np.random.normal(100.0,20.0,10000)

plt.hist(incomes,50)
plt.show()
```

```python
# desvio padrao raiz da variancia
incomes.std()
```

```python
# variancia media dos quadrados das diferencas pra media
incomes.var()
```
