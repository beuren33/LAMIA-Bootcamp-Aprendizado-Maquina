```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# tira a media de cada valor usado dentro da covariancia
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]

# formula da covariancia
def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1)

# gerando dados aleatorios
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

plt.scatter(pageSpeeds, purchaseAmount)
plt.xlabel('Page Speed')
plt.ylabel('Purchase Amount')
plt.show()

# confere com a funcao pronta do numpy
np.cov(pageSpeeds, purchaseAmount)
```

```python
# agora criando uma dependencia entre as variaveis, dividindo pelo pageSpeeds
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

covariance(pageSpeeds, purchaseAmount)
```

```python
# correlacao normaliza a covariancia entre -1 e 1
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy 

correlation(pageSpeeds, purchaseAmount)
```

```python
# comparando com o corrcoef do numpy
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
# relacao linear
purchaseAmount = 100 - pageSpeeds * 3

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

correlation(pageSpeeds, purchaseAmount)
# correlacao alta
```

```python
# repetindo a divisao para ver se a covariancia muda
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

covariance(pageSpeeds, purchaseAmount)
```
